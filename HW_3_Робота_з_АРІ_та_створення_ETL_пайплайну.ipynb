–í–∏ —Ç–∞–∫–æ–∂ –º–æ–∂–µ—Ç–µ –Ω–∞–ª–∞—à—Ç—É–≤–∞—Ç–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏ –ø—ñ–¥–∫–ª—é—á–µ–Ω–Ω—è –¥–æ –ë–î –±–µ–∑ .env —Ñ–∞–π–ª–∞, –ø—Ä–æ—Å—Ç–æ –ø—Ä–æ–ø–∏—Å–∞–≤—à–∏ —Ç–µ–∫—Å—Ç–æ–º –≤ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏—Ö –º—ñ—Å—Ü—è—Ö. –¶–µ - –Ω–µ —Ä–µ–∫–æ–º–µ–Ω–¥–æ–≤–∞–Ω–∏–π –ø—ñ–¥—Ö—ñ–¥.

````python
# –í—Å—Ç–∞–Ω–æ–≤–ª—é—î–º–æ —Å—É—á–∞—Å–Ω–∏–π —Å—Ç–µ–∫ –¥–ª—è —Ä–æ–±–æ—Ç–∏ –∑ –ë–î
!pip install sqlalchemy pymysql openpyxl requests python-dotenv 
````

````python
import datetime
import requests
import json
import os

from dotenv import load_dotenv
import pandas as pd
import sqlalchemy as sa
from sqlalchemy import create_engine, text, MetaData, Table
from sqlalchemy.orm import sessionmaker
````

````python
def create_connection():
    """
    –°—Ç–≤–æ—Ä—é—î –ø—ñ–¥–∫–ª—é—á–µ–Ω–Ω—è —á–µ—Ä–µ–∑ SQLAlchemy
    """
    # –ó–∞–≤–∞–Ω—Ç–∞–∂—É—î–º–æ –∑–º—ñ–Ω–Ω—ñ —Å–µ—Ä–µ–¥–æ–≤–∏—â–∞
    load_dotenv()

    # –û—Ç—Ä–∏–º—É—î–º–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏ –∑ environment variables
    host = os.getenv('DB_HOST', 'localhost')
    port = os.getenv('DB_PORT', '3306')
    user = os.getenv('DB_USER')
    password = os.getenv('DB_PASSWORD')
    database = os.getenv('DB_NAME')

    if not all([user, password, database]):
        raise ValueError("–ù–µ –≤—Å—ñ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏ –ë–î –∑–∞–¥–∞–Ω—ñ –≤ .env —Ñ–∞–π–ª—ñ!")

    # –°—Ç–≤–æ—Ä—é—î–º–æ connection string
    connection_string = f"mysql+pymysql://{user}:{password}@{host}:{port}/{database}"

    # –°—Ç–≤–æ—Ä—é—î–º–æ engine –∑ connection pooling
    engine = create_engine(
        connection_string,
        pool_size=2,           # –†–æ–∑–º—ñ—Ä –ø—É–ª—É –ø—ñ–¥–∫–ª—é—á–µ–Ω—å
        max_overflow=20,        # –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –¥–æ–¥–∞—Ç–∫–æ–≤–∏—Ö –ø—ñ–¥–∫–ª—é—á–µ–Ω—å
        pool_pre_ping=True,     # –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ –ø—ñ–¥–∫–ª—é—á–µ–Ω–Ω—è –ø–µ—Ä–µ–¥ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º
        echo=False              # –õ–æ–≥—É–≤–∞–Ω–Ω—è SQL –∑–∞–ø–∏—Ç—ñ–≤ (True –¥–ª—è debug)
    )

    # –¢–µ—Å—Ç—É—î–º–æ –ø—ñ–¥–∫–ª—é—á–µ–Ω–Ω—è
    try:
        with engine.connect() as conn:
            result = conn.execute(text("SELECT 1"))
            result.fetchone()

        print("‚úÖ –ü—ñ–¥–∫–ª—é—á–µ–Ω–Ω—è –¥–æ –ë–î —É—Å–ø—ñ—à–Ω–µ!")
        print(f"üîó {user}@{host}:{port}/{database}")
        print(f"‚ö° Engine: {engine}")

        return engine

    except Exception as e:
        print(f"‚ùå –ü–æ–º–∏–ª–∫–∞ –ø—ñ–¥–∫–ª—é—á–µ–Ω–Ω—è: {e}")
        return None

# –°—Ç–≤–æ—Ä—é—î–º–æ –ø—ñ–¥–∫–ª—é—á–µ–Ω–Ω—è
engine = create_connection()
````

### –ó–∞–≤–¥–∞–Ω–Ω—è 1: –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ç–∞–±–ª–∏—Ü—ñ –∫—É—Ä—Å—ñ–≤ –≤–∞–ª—é—Ç —Ç–∞ API —ñ–Ω—Ç–µ–≥—Ä–∞—Ü—ñ—è (2 –±–∞–ª–∏)

**–ü–æ–≤—Ç–æ—Ä—ñ—Ç—å –ø—Ä–æ—Ü–µ–¥—É—Ä—É –∑ –ª–µ–∫—Ü—ñ—ó:** —Å—Ç–≤–æ—Ä—ñ—Ç—å —Ç–∞–±–ª–∏—Ü—é –¥–ª—è –∫—É—Ä—Å—ñ–≤ –≤–∞–ª—é—Ç, –∞–ª–µ –≤–∂–µ –≤ —Ü—ñ–π –±–∞–∑—ñ –¥–∞–Ω–∏—Ö. –†–µ–∑—É–ª—å—Ç–∞—Ç–æ–º –º–∞—î –±—É—Ç–∏ –Ω–æ–≤–∞ —Ç–∞–±–ª–∏—Ü—è –∑ –∫—É—Ä—Å–∞–º–∏ –≤–∞–ª—é—Ç USD, EUR, UAH –≤ –ë–î (–º–æ–∂–Ω–∞ –∑–∞–≤–∞–Ω—Ç–∞–∂–∏—Ç–∏ –±—ñ–ª—å—à–µ –≤–∞–ª—é—Ç). –ü—Ä–æ–¥–µ–º–æ–Ω—Å—Ç—Ä—É–π—Ç–µ, —â–æ —Ç–∞–±–ª–∏—Ü—è –±—É–ª–∞ –¥–æ–¥–∞–Ω–∞, –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ SELECT.

–¢–æ–±—Ç–æ —Ç—É—Ç –≤–∏ –º–æ–∂–µ—Ç–µ –ø—Ä—è–º–æ —Å–∫–æ–ø—ñ—é–≤–∞—Ç–∏ –∫–æ–¥ –∑ –ª–µ–∫—Ü—ñ—ó, –≤–Ω–µ—Å—Ç–∏ –Ω–µ–æ–±—Ö—ñ–¥–Ω—ñ –∑–º—ñ–Ω–∏ —ñ –∑–∞–ø—É—Å—Ç–∏—Ç–∏. –ì–æ–ª–æ–≤–Ω–µ - –æ—Ç—Ä–∏–º–∞—Ç–∏ —Ç–∞–±–ª–∏—Ü—é –≤ –ë–î classicmodels.

````python
# –°—Ç–≤–æ—Ä—é—î–º–æ —Ç–∞–±–ª–∏—Ü—é –¥–ª—è –∫—É—Ä—Å—ñ–≤ –≤–∞–ª—é—Ç (—Å—É—á–∞—Å–Ω–∏–π —Å–ø–æ—Å—ñ–±)
def create_currency_table(engine):
    """–°—Ç–≤–æ—Ä—é—î —Ç–∞–±–ª–∏—Ü—é —á–µ—Ä–µ–∑ SQLAlchemy"""

    create_table_sql = text("""
    CREATE TABLE IF NOT EXISTS curr_rates (
        id INT AUTO_INCREMENT PRIMARY KEY,
        currency_code VARCHAR(3) NOT NULL,
        rate_to_usd DECIMAL(10, 6) NOT NULL,
        rate_date DATE NOT NULL,
        created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        updated_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP ON UPDATE CURRENT_TIMESTAMP,
        INDEX idx_currency_date (currency_code, rate_date),
        UNIQUE KEY unique_currency_date (currency_code, rate_date)
    )
    """)

    with engine.connect() as conn:
        # —Ç—É—Ç –ª–∏—à–µ –æ–¥–Ω–∞ —Ç—Ä–∞–Ω–∑–∞–∫—Ü—ñ—è - conn.begin() –Ω–µ —Ç—Ä–µ–±–∞
        conn.execute(create_table_sql)

    print("‚úÖ –¢–∞–±–ª–∏—Ü—è curr_rates —Å—Ç–≤–æ—Ä–µ–Ω–∞")

def fetch_exchange_rates():
    """–û—Ç—Ä–∏–º—É—î –∫—É—Ä—Å–∏ –≤–∞–ª—é—Ç –∑ API"""
    try:
        # –ë–µ–∑–∫–æ—à—Ç–æ–≤–Ω–∏–π API –∫—É—Ä—Å—ñ–≤ –≤–∞–ª—é—Ç
        url = "https://api.exchangerate-api.com/v4/latest/USD"
        response = requests.get(url, timeout=10)
        response.raise_for_status()

        data = response.json()

        # –í–∞–ª—é—Ç–∏ —â–æ –Ω–∞—Å —Ü—ñ–∫–∞–≤–ª—è—Ç—å
        currencies = ['EUR', 'GBP', 'UAH', 'PLN', 'JPY']
        rates = {}

        for currency in currencies:
            if currency in data['rates']:
                # –ó–±–µ—Ä—ñ–≥–∞—î–º–æ –∫—É—Ä—Å –¥–æ USD (—Å–∫—ñ–ª—å–∫–∏ –æ–¥–∏–Ω–∏—Ü—å –≤–∞–ª—é—Ç–∏ –∑–∞ 1 USD)
                rates[currency] = data['rates'][currency]

        return rates, datetime.date.today()

    except Exception as e:
        print(f"‚ùå –ü–æ–º–∏–ª–∫–∞ API: {e}")
        return None, None

def save_exchange_rates(engine, rates_dict, rate_date):
    """–ó–±–µ—Ä—ñ–≥–∞—î –∫—É—Ä—Å–∏ –≤ –ë–î –∑ –æ–±—Ä–æ–±–∫–æ—é –∫–æ–Ω—Ñ–ª—ñ–∫—Ç—ñ–≤"""

    if not rates_dict:
        print("‚ùå –ù–µ–º–∞—î –¥–∞–Ω–∏—Ö –¥–ª—è –∑–±–µ—Ä–µ–∂–µ–Ω–Ω—è")
        return False

    # SQL –∑ ON DUPLICATE KEY UPDATE –¥–ª—è MySQL
    insert_sql = text("""
    INSERT INTO curr_rates (currency_code, rate_to_usd, rate_date)
    VALUES (:currency, :rate, :date)
    ON DUPLICATE KEY UPDATE
        rate_to_usd = VALUES(rate_to_usd),
        updated_at = CURRENT_TIMESTAMP
    """)

    try:
        with engine.connect() as conn:
            with conn.begin():  # –¢—Ä–∞–Ω–∑–∞–∫—Ü—ñ—è –¥–ª—è –≤—Å—ñ—Ö –≤—Å—Ç–∞–≤–æ–∫
                for currency, rate in rates_dict.items():
                    conn.execute(insert_sql, {
                        'currency': currency,
                        'rate': rate,
                        'date': rate_date
                    })

        print(f"‚úÖ –ó–±–µ—Ä–µ–∂–µ–Ω–æ {len(rates_dict)} –∫—É—Ä—Å—ñ–≤ –≤–∞–ª—é—Ç –Ω–∞ {rate_date}")
        return True

    except Exception as e:
        print(f"‚ùå –ü–æ–º–∏–ª–∫–∞ –∑–±–µ—Ä–µ–∂–µ–Ω–Ω—è: {e}")
        return False
````

````python
# –í–∏–∫–æ–Ω—É—î–º–æ –ø–æ–≤–Ω–∏–π —Ü–∏–∫–ª API ‚Üí –ë–î
create_currency_table(engine)

print("üì° –û—Ç—Ä–∏–º—É—î–º–æ –∫—É—Ä—Å–∏ –≤–∞–ª—é—Ç...")
rates, date = fetch_exchange_rates()

if rates:
    print(f"–û—Ç—Ä–∏–º–∞–Ω—ñ –∫—É—Ä—Å–∏ –Ω–∞ {date}:")
    for currency, rate in rates.items():
        print(f"  1 USD = {rate:.4f} {currency}")

    # –ó–±–µ—Ä—ñ–≥–∞—î–º–æ –≤ –ë–î
    if save_exchange_rates(engine, rates, date):
        # –ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ –∑–±–µ—Ä–µ–∂–µ–Ω—ñ –¥–∞–Ω—ñ
        verification_df = pd.read_sql(
            "SELECT * FROM currency_rates ORDER BY created_at DESC LIMIT 10",
            engine
        )
        print("\n–ó–±–µ—Ä–µ–∂–µ–Ω—ñ –¥–∞–Ω—ñ:")
        display(verification_df)
````

### –Ø–∫ –ø–µ—Ä–µ–≤—ñ—Ä–∏—Ç–∏ —â–æ –≤—Å–µ –ø—Ä–∞—Ü—é—î:

- –í–∏–≤–µ–¥—ñ—Ç—å –∫—ñ–ª—å–∫—ñ—Å—Ç—å –∑–∞–ø–∏—Å—ñ–≤ –ø—ñ—Å–ª—è –∫–æ–∂–Ω–æ–≥–æ –∫—Ä–æ–∫—É
- –ü–æ–∫–∞–∂—ñ—Ç—å –ø–µ—Ä—à—ñ 5 —Ä—è–¥–∫—ñ–≤ –∫–æ–∂–Ω–æ—ó –∞–Ω–∞–ª—ñ—Ç–∏—á–Ω–æ—ó —Ç–∞–±–ª–∏—Ü—ñ

````python
def create_report(engine, output_dir="reports"):
    """
    –°—Ç–≤–æ—Ä—é—î –∫–æ–º–ø–ª–µ–∫—Å–Ω–∏–π –∞–Ω–∞–ª—ñ—Ç–∏—á–Ω–∏–π –∑–≤—ñ—Ç –∑ ETL –ø—Ä–æ—Ü–µ—Å–æ–º
    """
    print("üöÄ –ó–∞–ø—É—Å–∫ ETL –ø–∞–π–ø–ª–∞–π–Ω—É...")

    # –°—Ç–≤–æ—Ä—é—î–º–æ –¥–∏—Ä–µ–∫—Ç–æ—Ä—ñ—é –¥–ª—è –∑–≤—ñ—Ç—ñ–≤
    os.makedirs(output_dir, exist_ok=True)
    timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
    
    try:
        # EXTRACT: –í–∏—Ç—è–≥—É—î–º–æ –¥–∞–Ω—ñ –∑ –∫—ñ–ª—å–∫–æ—Ö –¥–∂–µ—Ä–µ–ª

        print("üì• 1. EXTRACT - –í–∏—Ç—è–≥—É–≤–∞–Ω–Ω—è –¥–∞–Ω–∏—Ö...")

        # 1.1 –¥–∞–Ω—ñ –ø—Ä–æ –≤–∏–∫–æ–Ω–∞–Ω—ñ –∑–∞–º–æ–≤–ª–µ–Ω–Ω—è –∑–∞ 2004 —Ä—ñ–∫ - –∑'—î–¥–Ω–∞—Ç–∏ —Ç–∞–±–ª–∏—Ü—ñ orders, orderdetails, products, customers
        orders_query = text("""
        SELECT
        o.orderNumber,
        o.orderDate,
        c.customerName,
        c.city,
        c.state, 
        c.country,
        p.productName,
        p.buyPrice ,
        p.quantityInStock,
        od.quantityOrdered,
        od.priceEach
    FROM
        orders AS o
    JOIN
        customers AS c ON o.customerNumber = c.customerNumber
    JOIN
        orderdetails AS od ON o.orderNumber = od.orderNumber
    JOIN
        products AS p ON od.productCode = p.productCode
    WHERE
        YEAR(o.orderDate) = 2004
        AND o.status = 'Shipped'
    ORDER BY
        o.orderDate
        """)

        df_orders = pd.read_sql(orders_query, engine) #, parse_dates=['hire_date', 'birth_date'])

                # 1.2 –¥–∞–Ω—ñ –ø—Ä–æ –ø—Ä–æ–¥—É–∫—Ç–∏ - –Ω–∞–∑–≤–∏, –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó, —Ü—ñ–Ω–∏
        product_query = text("""
         SELECT
    p.productName,
    pl.productLine,
    p.buyPrice
FROM
    products AS p
JOIN
    productlines AS pl ON p.productLine = pl.productLine
ORDER BY
    pl.productLine, p.productName;
        """)

        df_product = pd.read_sql(product_query, engine) #, parse_dates=['from_date', 'to_date'])
        
         # 1.3 –ö—É—Ä—Å–∏ –≤–∞–ª—é—Ç
        try:
            df_currencies = pd.read_sql("SELECT * FROM curr_rates ORDER BY rate_date DESC", engine)
            eur_rate = df_currencies.loc[df_currencies['currency'] == 'EUR', 'rate'].iloc[0]
        except:
            df_currencies = pd.DataFrame()  # –Ø–∫—â–æ —Ç–∞–±–ª–∏—Ü—ñ –Ω–µ–º–∞—î

        print(f"–ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–æ: {len(df_orders)} –∑–∞–ø–∏—Å—ñ–≤ –≤–∏–∫–æ–Ω–∞–Ω–∏—Ö –∑–∞–º–æ–≤–ª–µ–Ω—å, {len(df_product)} –∑–∞–ø–∏—Å—ñ–≤ –ø—Ä–æ–¥—É–∫—Ç—ñ–≤")

        # TRANSFORM: –û–±—Ä–æ–±–ª—è—î–º–æ —Ç–∞ –∑–±–∞–≥–∞—á—É—î–º–æ –¥–∞–Ω—ñ –≤ Pandas
        print("üîß 2. TRANSFORM - –û–±—Ä–æ–±–∫–∞ —Ç–∞ –∑–±–∞–≥–∞—á–µ–Ω–Ω—è...")
        # profit_per_item - –ø—Ä–∏–±—É—Ç–æ–∫ –∑ –æ–¥–Ω–æ–≥–æ —Ç–æ–≤–∞—Ä—É (–≤–∏–∫–æ—Ä–∏—Å—Ç–∞–π—Ç–µ –∫–æ–ª–æ–Ω–∫–∏: priceEach - buyPrice)
        df_orders['profit_per_item'] = df_orders['priceEach'] - df_orders['buyPrice']
                
        # total_profit - –∑–∞–≥–∞–ª—å–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ –∑ —Ç–æ–≤–∞—Ä–Ω–æ—ó –ø–æ–∑–∏—Ü—ñ—ó (–≤–∏–∫–æ—Ä–∏—Å—Ç–∞–π—Ç–µ –∫–æ–ª–æ–Ω–∫–∏: profit_per_item √ó quantityOrdered)
        df_orders['total_profit'] = df_orders['profit_per_item'] * df_orders['quantityOrdered']
        
        #total_amount_eur - —Å—É–º–∞ –≤ —î–≤—Ä–æ (–≤–∏–∫–æ—Ä–∏—Å—Ç–∞–π—Ç–µ –∫–æ–ª–æ–Ω–∫–∏: total_amount / eur_rate)
        df_orders['total_amount'] = df_orders['priceEach'] * df_orders['quantityOrdered']
        df_orders['total_amount_eur'] = df_orders['total_amount'] / eur_rate

        #2.2 –°—Ç–≤–æ—Ä–∏—Ç–∏ –∞–Ω–∞–ª—ñ—Ç–∏—á–Ω—É —Ç–∞–±–ª–∏—Ü—é –ø–æ –∫—Ä–∞—ó–Ω–∞—Ö (–¢–û–ü-5):

        country_agg = df_orders.groupby('country').agg(
            unique_orders=('orderNumber', 'nunique'), #–ö—ñ–ª—å–∫—ñ—Å—Ç—å —É–Ω—ñ–∫–∞–ª—å–Ω–∏—Ö –∑–∞–º–æ–≤–ª–µ–Ω—å - —É–Ω—ñ–∫–∞–ª—å–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è –∫–æ–ª–æ–Ω–∫–∏ orderNumber
            total_amount=('total_amount', 'sum'),     #–ó–∞–≥–∞–ª—å–Ω–∏–π –¥–æ—Ö—ñ–¥ - —Å—É–º–∞ –∫–æ–ª–æ–Ω–∫–∏ total_amount
            total_profit=('total_profit', 'sum'),     #–ó–∞–≥–∞–ª—å–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ - —Å—É–º–∞ –∫–æ–ª–æ–Ω–∫–∏ total_profit
            total_quantity=('quantityOrdered', 'sum')  #–ö—ñ–ª—å–∫—ñ—Å—Ç—å –ø—Ä–æ–¥–∞–Ω–∏—Ö —Ç–æ–≤–∞—Ä—ñ–≤ - —Å—É–º–∞ –∫–æ–ª–æ–Ω–∫–∏ quantityOrdered
        ).reset_index()
        
        #–ú–∞—Ä–∂–∞ –ø—Ä–∏–±—É—Ç–∫—É (%) - (–∑–∞–≥–∞–ª—å–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ / –∑–∞–≥–∞–ª—å–Ω–∏–π –¥–æ—Ö—ñ–¥) √ó 100
        country_agg['profit_margin_percent'] = (country_agg['total_profit'] / country_agg['total_amount']) * 100 
        df_countries_top5 = country_agg.sort_values(by='total_amount', ascending=False).head(5)

         # 2.3 –°—Ç–≤–æ—Ä–∏—Ç–∏ –∞–Ω–∞–ª—ñ—Ç–∏—á–Ω—ñ —Ç–∞–±–ª–∏—Ü—ñ –ø–æ –ø—Ä–æ–¥—É–∫—Ç–æ–≤–∏—Ö –ª—ñ–Ω—ñ—è—Ö
        
        product_agg = df_orders.groupby('productName').agg(
            unique_orders=('orderNumber', 'nunique'), #–ö—ñ–ª—å–∫—ñ—Å—Ç—å —É–Ω—ñ–∫–∞–ª—å–Ω–∏—Ö –∑–∞–º–æ–≤–ª–µ–Ω—å - —É–Ω—ñ–∫–∞–ª—å–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è –∫–æ–ª–æ–Ω–∫–∏ orderNumber
            total_amount=('total_amount', 'sum'),      #–ó–∞–≥–∞–ª—å–Ω–∏–π –¥–æ—Ö—ñ–¥ - —Å—É–º–∞ –∫–æ–ª–æ–Ω–∫–∏ total_amount
            total_profit=('total_profit', 'sum'),      #–ó–∞–≥–∞–ª—å–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ - —Å—É–º–∞ –∫–æ–ª–æ–Ω–∫–∏ total_profit
            total_quantity=('quantityOrdered', 'sum')  #–ö—ñ–ª—å–∫—ñ—Å—Ç—å –ø—Ä–æ–¥–∞–Ω–∏—Ö —Ç–æ–≤–∞—Ä—ñ–≤ - —Å—É–º–∞ –∫–æ–ª–æ–Ω–∫–∏ quantityOrdered
        ).reset_index()

        product_agg['profit_margin_percent'] = (product_agg['total_profit'] / product_agg['total_amount']) * 100  #–ú–∞—Ä–∂–∞ –ø—Ä–∏–±—É—Ç–∫—É (%) - (–∑–∞–≥–∞–ª—å–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ / –∑–∞–≥–∞–ª—å–Ω–∏–π –¥–æ—Ö—ñ–¥) √ó 100
        df_product_lines = product_agg.sort_values(by='total_amount', ascending=False)

        # 3.1 –ó–±–µ—Ä–µ–∂–µ–Ω–Ω—è Excel —Ñ–∞–π–ª—É –∑ –≤–∫–ª–∞–¥–∫–∞–º–∏
        output_file_path = os.path.join(output_dir, f"sales_report_{timestamp}.xlsx")
        
        with pd.ExcelWriter(output_file_path, engine='xlsxwriter') as writer:
            df_countries_top5.to_excel(writer, sheet_name='Top_Countries', index=False)
            df_product_lines.to_excel(writer, sheet_name='Product_Lines', index=False)
            print("‚úÖ –ó–≤—ñ—Ç —É—Å–ø—ñ—à–Ω–æ –∑–±–µ—Ä–µ–∂–µ–Ω–æ —É —Ñ–∞–π–ª Excel.")

````
